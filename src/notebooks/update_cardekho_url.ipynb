# Enhance AI Car Matchmaker

This notebook updates `cars.json` with a new `cardekhoUrl` field for every car, updates the TypeScript `Car` type, and verifies the recommendation card changes.

```python
from pathlib import Path
import json

cars_path = Path("src/data/cars.json")
with cars_path.open("r", encoding="utf-8") as f:
    cars = json.load(f)

print("First car before update:")
print(json.dumps(cars[0], indent=2)[:400])
print(f"Total cars: {len(cars)}")
```

```python
def slug(value: str) -> str:
    value = value.lower()
    value = ''.join(c if c.isalnum() or c == ' ' or c == '-' else '' for c in value)
    value = '-'.join(value.split())
    while '--' in value:
        value = value.replace('--', '-')
    return value

cars_path = Path("src/data/cars.json")
with cars_path.open("r", encoding="utf-8") as f:
    cars = json.load(f)

for car in cars:
    brand_slug = slug(car['brand'])
    model_slug = slug(car['model'])
    car['cardekhoUrl'] = f"https://www.cardekho.com/{brand_slug}/{model_slug}"

with cars_path.open("w", encoding="utf-8") as f:
    json.dump(cars, f, indent=2, ensure_ascii=False)
    f.write("\n")

print("Updated cars.json with cardekhoUrl for all cars.")
```
